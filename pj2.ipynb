```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  
# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df['month_posted_date'] = df.job_posted_date.dt.strftime('%b')
```

```python
months = df.month_posted_date.unique()
months
```

```python
df_month = {month : df[df['month_posted_date'] == month] for month in months}
```

```python
df_month['Jan']
```

```python
df_qtr = pd.concat([df_month['Jan'],df_month['Feb'],df_month['Mar'] ], ignore_index=True)
```

```python
df_qtr['month_posted_date'].value_counts().plot(kind='bar') 
```

```python
df_qtr.plot(kind ='line')
```
